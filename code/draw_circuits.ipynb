```python
import qiskit
from qiskit import QuantumCircuit
```

```python
from qiskit.circuit.library import ZFeatureMap, ZZFeatureMap
import matplotlib.pyplot as plt
```

```python
# 2-qubit ZFeatureMap, 2 repetitions
z_fm = ZZFeatureMap(feature_dimension=2, reps=2)

# Decompose into basic one-qubit gates
z_fm_decomp = z_fm.decompose()

fig = z_fm_decomp.draw(output="mpl")   # fold=-1 keeps it on one line

# Save as high-quality PDF (vector graphic)
fig.savefig("zz_decomposed.pdf", format="pdf", dpi=600, bbox_inches="tight")

plt.close(fig)
```
